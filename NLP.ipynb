```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
nltk.download('stopwords')
```

```python
text = "Hello!!! I am LearNing Data ScIENce and loving it."

# Lowercase
text = text.lower()
print(text)

 

# Remove punctuation
text = text.translate(str.maketrans('','',string.punctuation))
print(text)

 

# split into words(tokenization)

words = text.split()

print(words)

 

# remove stopwords

stop_words = set(stopwords.words('english'))

words = [word for word in words if word not in stop_words]

print(words)

 

# Stemming

stemmer = PorterStemmer()

words = [stemmer.stem(word) for word in words]

print(words)

 

# Finl clean text

clean_text = " ".join(words)

print(clean_text)
```

```python
text = "Wow!!! Data-Science is Awesome. I've been learning it since 2023, and honestly, It's super-interesting!!!" #AI #ML

# Lowercase
text = text.lower()
print(text)

# Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))
print(text)

# Tokenization
words = text.split()
print(words)

# Remove stopwords
stop_words = set(stopwords.words('english'))
words = [word for word in words if word not in stop_words]
print(words)

# Stemming
stemmer = PorterStemmer()
words = [stemmer.stem(word) for word in words]
print(words)

# Final clean text
clean_text = " ".join(words)
print(clean_text)
```
