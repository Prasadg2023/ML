```python
# Step 1: Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 2: Create sample data
data = {
    'Experience': [1, 2, 3, 4, 5],
    'Salary': [30000, 40000, 50000, 60000, 70000]
}

df = pd.DataFrame(data)

# Step 3: Define X and y
X = df[['Experience']]   # independent variable
y = df['Salary']         # dependent variable

# Step 4: Create model
model = LinearRegression()

# Step 5: Train model
model.fit(X, y)

# Step 6: Predict
predicted_salary = model.predict([[6]])

print("Predicted Salary for 6 years experience:", predicted_salary[0])
```

```python
# Import libraries
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# -------------------------------
# Sample Dataset (Hours vs Marks)
# -------------------------------
hours = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
marks = np.array([35, 40, 50, 55, 65, 70, 80, 90])

# Train Model
model = LinearRegression()
model.fit(hours, marks)

# -------------------------------
# UI Design
# -------------------------------
st.title("📊 Predict Marks")

# Input box (like your image)
study_hours = st.number_input("Enter Study Hours", min_value=0.0, max_value=24.0, step=0.5)

# Predict Button
if st.button("Predict"):
    prediction = model.predict([[study_hours]])
    st.success(f"Predicted Marks: {prediction[0]:.2f}")

# -------------------------------
# Graph Section
# -------------------------------
st.subheader("Graph")

# Plot
plt.figure()
plt.scatter(hours, marks)
plt.plot(hours, model.predict(hours))
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")

st.pyplot(plt)
```

```python
import streamlit as st

import pandas as pd

import matplotlib.pyplot as plt

import numpy as np

from sklearn.linear_model import LinearRegression

st.title("Student Marks Prediction using Linear Regression")

data = {

    "Study Hours" : [1,2,3,4,5,6],

    "Marks":[35,45,55,65,75,85]

}

df = pd.DataFrame(data)

st.write(df)

# Train Model

X = df[["Study Hours"]]

y = df["Marks"]

model = LinearRegression()

model.fit(X,y)

# Prediction

st.subheader("Predict Marks")

hours = st.number_input("Enter Study Hours", min_value = 0.0, max_value = 12.0, value = 3.5)

if st.button("Predict"):

    prediction = model.predict([[hours]])

    st.success(f"Predicted Marks: {prediction[0]:.2f}")

# Graph

st.subheader("Graph")

plt.scatter(df["Study Hours"], df["Marks"])

x_range = np.linspace(0, 12, 100)

y_range = model.predict(x_range.reshape(-1,1))

plt.plot(x_range, y_range)

plt.xlabel("Study Hours")

plt.ylabel("Marks")

st.pyplot(plt)
```

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# -------------------------------
# Create Dataset
# -------------------------------
data = {
    "Study Hours": [1, 2, 3, 4, 5, 6],
    "Marks": [35, 45, 55, 65, 75, 85]
}

df = pd.DataFrame(data)

print("Dataset:")
print(df)

# -------------------------------
# Train Model
# -------------------------------
X = df[["Study Hours"]]
y = df["Marks"]

model = LinearRegression()
model.fit(X, y)

# -------------------------------
# User Input
# -------------------------------
hours = float(input("\nEnter Study Hours: "))

# -------------------------------
# Prediction
# -------------------------------
prediction = model.predict([[hours]])
print(f"\nPredicted Marks: {prediction[0]:.2f}")

# -------------------------------
# Model Accuracy
# -------------------------------
y_pred = model.predict(X)
print(f"R2 Score: {r2_score(y, y_pred):.2f}")

# -------------------------------
# Graph
# -------------------------------
plt.figure(figsize=(6,4))

# Scatter plot
plt.scatter(df["Study Hours"], df["Marks"], label="Actual Data")

# Regression line
x_range = np.linspace(0, 12, 100)
y_range = model.predict(x_range.reshape(-1,1))

plt.plot(x_range, y_range, label="Regression Line")

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.legend()

plt.show()
```
